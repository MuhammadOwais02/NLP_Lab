```python
# import necessary libraries
import pandas as pd
import demoji
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import string
```

## Preprocessing

```python
# define a function to preprocess the text
def preprocess_text(text):
    # remove emoji
    text = demoji.replace(text, "")
    # remove URLs
    text = " ".join(word for word in text.split() if not "http" in word)
    # remove usernames
    text = " ".join(word for word in text.split() if not word.startswith("@"))
    # remove punctuation
    text = "".join(word for word in text if not word in string.punctuation)
    # convert to lowercase
    text = text.lower()
    return text
```

## Sentiment Analysis

TextBlob

```python
# define a function to calculate the sentiment using TextBlob
def get_textblob_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
```

VANDER

```python
# define a function to calculate the sentiment using VADER
def get_vader_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(text)
    return scores["compound"]
```

Final Sentiments: 

```python
# apply preprocessing and print the results
for index, row in df.iterrows():
    preprocessed_text = preprocess_text(row[0])
    textblob_sentiment = get_textblob_sentiment(preprocessed_text)
    vader_sentiment = get_vader_sentiment(preprocessed_text)
    print(f"TextBlob sentiment of tweet {index+1}: {textblob_sentiment}")
    print(f"VADER sentiment of tweet {index+1}: {vader_sentiment}")
```

```python
# apply preprocessing and add new columns for preprocessed text, TextBlob sentiment, and VADER sentiment
df['preprocessed_text'] = df['tweet'].apply(preprocess_text)
df['textblob_sentiment'] = df['preprocessed_text'].apply(get_textblob_sentiment)
df['vader_sentiment'] = df['preprocessed_text'].apply(get_vader_sentiment)
```

```python
def get_sentiment(row):
    if row['textblob_sentiment'] > 0 and row['vader_sentiment'] > 0:
        return 'positive'
    elif row['textblob_sentiment'] < 0 and row['vader_sentiment'] < 0:
        return 'negative'
    else:
        return 'neutral'
```

## Models report

TextBlob + TF-IDF + LinearSVC

```python
from textblob import TextBlob
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define function to extract TextBlob sentiment
def get_textblob_sentiment(text):
    sentiment = TextBlob(text).sentiment.polarity
    return 'positive' if sentiment > 0 else 'negative' if sentiment < 0 else 'neutral'

# Apply TextBlob sentiment analysis to the preprocessed text
df['textblob_sentiment'] = df['preprocessed_text'].apply(get_textblob_sentiment)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['preprocessed_text'], df['textblob_sentiment'], test_size=0.2, random_state=42)

# Define the TF-IDF vectorizer and transform the data
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Define the LinearSVC model and fit the data
model = LinearSVC()
model.fit(X_train_tfidf, y_train)

# Make predictions and evaluate the model
y_pred = model.predict(X_test_tfidf)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred, average='weighted')}")
print(f"Recall: {recall_score(y_test, y_pred, average='weighted')}")
print(f"F1 score: {f1_score(y_test, y_pred, average='weighted')}")
```

TextBlob + CountVect + TF-IDF + Logistic Regression

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from textblob import TextBlob
import scipy


# add TextBlob sentiment column
df['textblob_sentiment'] = df['preprocessed_text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# create count vectorizer and tf-idf vectorizer
count_vectorizer = CountVectorizer()
tfidf_vectorizer = TfidfVectorizer()

# fit and transform the text data using count vectorizer
X_count = count_vectorizer.fit_transform(df['preprocessed_text'])
# fit and transform the text data using tf-idf vectorizer
X_tfidf = tfidf_vectorizer.fit_transform(df['preprocessed_text'])

# create the feature matrix by combining the TextBlob sentiment with the count vectorizer and tf-idf vectorizer
X_count_tfidf = scipy.sparse.hstack([X_count, X_tfidf, df['textblob_sentiment'].to_numpy().reshape(-1, 1)])

# set the target variable
y = df['sentiment']

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_count_tfidf, y, test_size=0.3, random_state=42)

# create the logistic regression model
logreg = LogisticRegression()

# train the model on the training set
logreg.fit(X_train, y_train)

# predict the sentiment for the test set
y_pred = logreg.predict(X_test)

# evaluate the performance of the model
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='weighted')
rec = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy : {acc:.2f}")
print(f"Precision: {prec:.2f}")
print(f"Recall   : {rec:.2f}")
print(f"F1 Score : {f1:.2f}")
```
