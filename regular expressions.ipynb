#### Regix patterns

Matching a specific word: r"word"

Matching any single character: r"."

Matching digits (0-9): r"\d"

Matching non-digits: r"\D"

Matching whitespace characters: r"\s"

Matching non-whitespace characters: r"\S"

Matching word characters (alphanumeric and underscore): r"\w"

Matching non-word characters: r"\W"

Matching specific characters in a character set: r"[aeiou]"

Matching characters not in a character set: r"[^aeiou]"

Matching specific character ranges: r"[a-z]"

Matching one or more occurrences of a pattern: r"pattern+"

Matching zero or more occurrences of a pattern: r"pattern*"

Matching zero or one occurrence of a pattern: r"pattern?"

Matching a specific number of occurrences of a pattern: r"pattern{3}"

Matching a range of occurrences of a pattern: r"pattern{2,4}"

Matching the beginning of a string: r"^start"

Matching the end of a string: r"end$"

Grouping and capturing parts of a pattern: r"(pattern1)|(pattern2)"

Using backreferences to refer to captured groups: r"(?P<name>pattern)\s(?P=name)"

```python
import re

# Define the regular expression pattern
pattern = r"\b\d{3}-\d{3}-\d{4}\b"  # Example pattern for matching phone numbers in xxx-xxx-xxxx format

# Sample text
text = "John's phone number is 123-456-7890. Please call him."

# Find all matches of the pattern in the text
matches = re.findall(pattern, text)

# Iterate over the matches
for match in matches:
    print("Found match:", match)
```

```python
import re

# Define the regular expression pattern
pattern = r"\b\d{3}-\d{3}-\d{4}\b"  # Example pattern for matching phone numbers in xxx-xxx-xxxx format

# Sample text
text = "John's phone number is 123-456-7890. Please call him."

# Define the replacement string
replacement = "[PHONE]"

# Substitute the matches with the replacement string
new_text = re.sub(pattern, replacement, text)

print("Original text:", text)
print("Modified text:", new_text)
```

```python
import re

# Define the regular expression pattern
pattern = r"\s"  # Example pattern for splitting based on whitespace

# Sample text
text = "Hello World! How are you?"

# Split the text based on the pattern
splitted = re.split(pattern, text)

print(splitted)  # Output: ['Hello', 'World!', 'How', 'are', 'you?']
```

```python
import re

# Define the regular expression pattern
pattern = r"\b\d{3}-\d{3}-\d{4}\b"  # Example pattern for matching phone numbers in xxx-xxx-xxxx format

# Sample text
text = "John's phone number is 123-456-7890. Please call him."

# Find the first occurrence of the pattern
match = re.search(pattern, text)

if match:
    print("First occurrence found:", match.group())  # Output: 123-456-7890
else:
    print("No match found.")
```

```python
import re

# Define the regular expression pattern
pattern = r"\b\d{3}-\d{3}-\d{4}\b"  # Example pattern for matching phone numbers in xxx-xxx-xxxx format

# Sample text
text = "John's phone number is 123-456-7890. Please call him."

# Check if the pattern matches the text
if re.match(pattern, text):
    print("Pattern matches the text.")
else:
    print("Pattern does not match the text.")
```
